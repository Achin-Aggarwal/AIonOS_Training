```python
import json
import tiktoken

import pandas as pd

from langchain.text_splitter import RecursiveCharacterTextSplitter

from langchain_community.document_loaders import PyPDFDirectoryLoader
from langchain_community.embeddings.sentence_transformer import (
    SentenceTransformerEmbeddings
)
from langchain_community.vectorstores import Chroma

from google.colab import userdata
```

```python
api_key = userdata.get('GROQ_API_KEY')
```

```python
client = Groq(api_key = api_key)
```

```python
model_name = 'openai/gpt-oss-20b' # deployment name
```

```python
embedding_model = SentenceTransformerEmbeddings(model_name='thenlper/gte-large')
```

```python
!unzip resume_db.zip
```

```python
tesla_10k_collection = 'tesla-10k-2019-to-2023'
```

```python
vectorstore_persisted = Chroma(
    collection_name=tesla_10k_collection,
    persist_directory='/content/resume_db',
    embedding_function=embedding_model
)
```

```python
retriever = vectorstore_persisted.as_retriever(
    search_type='similarity',
    search_kwargs={'k': 5}
)
```

```python
qna_system_message = """
You are an assistant that summarizes PDF documents stored in a vector database created from the folder named 'resume_db'.

User input will have the context required by you to generate the summaries.
This context will begin with the token: ###Context.
The context contains relevant text extracted from each PDF file.

User requests will begin with the token: ###Question.

When summarizing:
1. For each PDF file, start with the exact file name (including extension) as a heading.
2. Below the heading, provide a clear, concise, and comprehensive summary covering all major points from that PDF.
3. Keep the tone professional and easy to understand.



Output format:
### <file_name.pdf>
<summary here>
"""
```

```python
qna_user_message_template = """
###Context
Here are some documents that are relevant to the question mentioned below.
{context}

###Question
{question}
"""
```

```python
user_input = "Summarize the PDF."
```

```python
relevant_document_chunks = retriever.get_relevant_documents(user_input)
```

```python
len(relevant_document_chunks)
```

```python
for document in relevant_document_chunks:
    print(document.page_content.replace("\t", " "))
    break
```

```python
relevant_document_chunks = retriever.get_relevant_documents(user_input)
context_list = [d.page_content for d in relevant_document_chunks]
context_for_query = ". ".join(context_list)

prompt = [
    {'role':'system', 'content': qna_system_message},
    {'role': 'user', 'content': qna_user_message_template.format(
         context=context_for_query,
         question=user_input
        )
    }
]

try:
    response = client.chat.completions.create(
        model=model_name,
        messages=prompt,
        temperature=0
    )

    prediction = response.choices[0].message.content.strip()
except Exception as e:
    prediction = f'Sorry, I encountered the following error: \n {e}'

print(prediction)
```
